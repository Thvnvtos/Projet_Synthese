```python
import numpy as np
import matplotlib.pyplot as plt

scores, dv, dt0, dt1, dh = [], [], [], [], []

with open("logs.txt", 'r') as f:
    line = f.readline()
    while line:
        x = line.split()
        scores.append(float(x[0]))
        dv.append(float(x[1]))
        dt0.append(float(x[2]))
        dt1.append(float(x[3]))
        dh.append(float(x[4]))
        line = f.readline()
```

```python
dv = (dv - np.mean(dv))/np.std(dv)
dt0 = (dt0 - np.mean(dt0))/np.std(dt0)
dt1 = (dt1 - np.mean(dt1))/np.std(dt1)
dh = (dh - np.mean(dh))/np.std(dh)
```

```python
plt.figure(figsize=(10,10))
plt.plot(scores)
plt.title("Best solution score")
plt.figure(figsize=(10,10))
plt.plot(dv)
plt.plot(dt0)
plt.plot(dt1)
plt.plot(dh)
plt.legend(['dv','dt0','dt1','dh'])
plt.title("uncertainties")
```
